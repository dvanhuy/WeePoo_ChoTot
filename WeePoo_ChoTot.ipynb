```python
def getProduct_category(idcategories):
    productsReturn = []
    with open('data.json', 'r',encoding="utf-8") as json_file:
        dataload = json.load(json_file)
    for product in dataload:
        if (product["category"] == idcategories):
            print(product)
            productsReturn.append(product)
    return productsReturn
            
getProduct_category(8030)
```

```python
def getProduct_time(number,unit = 1):
    productsReturn = []
    with open('data.json', 'r',encoding="utf-8") as json_file:
        dataload = json.load(json_file)
    for product in dataload:
        if (number*unit >= string_to_number(product["date"])):
            productsReturn.append(product)
    return productsReturn
getProduct_time(20,60)
```

```python
def string_to_number(stringtime):
    parts = stringtime.strip().split()
    heso = 1
    if (parts[1] == 'phút'):
        heso = 60
    elif (parts[1] == 'giờ'):
        heso = 3600
    else:
        heso = 1
    return int(parts[0])*heso
```
